```python
import tensorflow as tf
import matplotlib.pyplot as plt
print(tf.__version__)
```

```python
x_train = tf.random.normal(shape=(1000, ), dtype=tf.float32)
y_train = 3*x_train + 1 + 0.2*tf.random.normal(shape=(1000, ), dtype=tf.float32)

x_test = tf.random.normal(shape=(100, ), dtype=tf.float32)
y_test = 3*x_test + 1 + 0.2*tf.random.normal(shape=(100, ), dtype=tf.float32)

w = tf.Variable(-1.)
b = tf.Variable(-1.)

LR = 0.01
EPOCHS = 10

w_trace, b_trace = [], []
```

```python

for i in range(EPOCHS):
    for x, y in zip(x_train, y_train):
        with tf.GradientTape() as tape:
            hypothesis = w*x + b
            loss = tf.reduce_mean(tf.square(y-hypothesis))
        w1, b1 = tape.gradient(loss, [w, b])
        w_trace.append(w.numpy())
        b_trace.append(b.numpy())
        w = tf.Variable(w - LR * w1)
        b = tf.Variable(b - LR * b1)

        
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(w_trace, label='weight')
ax.plot(b_trace, label='bias')

ax.tick_params(labelsize=20)
ax.legend(fontsize=30)
ax.grid
```

```python
x_train = tf.random.normal(shape=(1000, ), dtype=tf.float32)
y_train = 3*x_train + 1 + 0.2*tf.random.normal(shape=(1000, ), dtype=tf.float32)

model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1, input_shape=(1, )))

model.compile(optimizer='sgd', loss='mse')

model.fit(x_train, y_train, epochs=100)

# 모델 테스트
model.evaluate(x_test, y_test, verbose=2)
```
